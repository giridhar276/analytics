## 49. Write a function to automate the data cleaning process for this dataset.

```python
def clean_data(df):
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    df.fillna(df.mean(), inplace=True)
    return df

cleaned_df = clean_data(df.copy())
cleaned_df.head()
```

## 50. Create a new column 'Success_Predictor' using a simple linear model on 'Total_Score'.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = df[['Math_Score', 'Science_Score', 'English_Score', 'History_Score']]
y = df['Total_Score']
model.fit(X, y)
df['Success_Predictor'] = model.predict(X)
df.head()
```
