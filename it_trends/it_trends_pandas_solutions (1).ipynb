## 46. Create a pipeline to preprocess the 'Revenue' and 'Market_Share' columns.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('scaler', StandardScaler())
])
df[['Revenue', 'Market_Share']] = pipeline.fit_transform(df[['Revenue', 'Market_Share']])
df.head()
```

## 50. Write a function to automate the data cleaning process for this dataset.

```python
def clean_data(df):
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    df.fillna(df.mean(), inplace=True)
    return df

cleaned_df = clean_data(df.copy())
cleaned_df.head()
```
