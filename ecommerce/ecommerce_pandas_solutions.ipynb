## 1. Load the dataset and display the first 5 rows.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('ecommerce_dataset.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

## 45. Create a pipeline to preprocess the 'Total_Cost' and 'Price_per_Unit' columns.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

pipeline = Pipeline([
    ('scaler', StandardScaler())
])
df[['Total_Cost', 'Price_per_Unit']] = pipeline.fit_transform(df[['Total_Cost', 'Price_per_Unit']])
df.head()
```

## 49. Write a function to automate the data cleaning process for this dataset.

```python
def clean_data(df):
    df.dropna(inplace=True)
    df.drop_duplicates(inplace=True)
    df.fillna(df.mean(), inplace=True)
    return df

cleaned_df = clean_data(df.copy())
cleaned_df.head()
```
